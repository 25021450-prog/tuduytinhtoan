W2A3:Viết chương trình nhận vào hai số nguyên a, b từ bàn phím. In ra tổng, hiệu, tích, phần 
nguyên, phần dư và kết quả chia thực của phép chia a cho b

```python
a = int(input("Nhập số nguyên a: "))
b = int(input("Nhập số nguyên b: "))
tong = a + b
hieu = a - b
tich = a * b
phan_nguyen = a // b
phan_du = a % b
chia_thuc = a / b
print(f"Tổng: {tong}")
print(f"Hiệu: {hieu}")
print(f"Tích: {tich}")
print(f"Phần nguyên: {phan_nguyen}")
print(f"Phần dư: {phan_du}")
print(f"Chia thực: {chia_thuc:.2f}")
```

6 đầu điểm: a1, b1, c1, a2, b2, a3. Trong đó:
• Các điểm a1, b1, c1 được nhân hệ số 1,
• Các điểm a2, b2 được nhân hệ số 2,

```python
input_string = input()
point_list = input_string.split()
a1 = float(point_list[0])
b1 = float(point_list[1])
c1 = float(point_list[2])
a2 = float(point_list[3])
b2 = float(point_list[4])
a3 = float(point_list[5])
T = ((a1+b1+c1)+(a2+b2)*2+a3*3)/10
print(f"diem trung binh = {T}")
```
