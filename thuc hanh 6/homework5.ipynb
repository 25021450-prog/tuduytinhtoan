```python
#28

import math
def gocb(a,b,c):
    vo_huong=(a[0]-b[0])*(c[0]-b[0])+(a[1]-b[1])*(c[1]-b[1])
    do_dai=(math.sqrt((a[0]-b[0])**2+(a[1]-b[1])**2))*(math.sqrt((c[0]-b[0])**2+(c[1]-b[1])**2))
    x=vo_huong/do_dai
    return math.acos(x)*(180/math.pi)
a=tuple(list(map(int,input().split())))
b=tuple(list(map(int,input().split())))
c=tuple(list(map(int,input().split())))
gocb(a,b,c)
```

```python
#30
def to_hop_chap(k,n):
    ng=1
    for i in range(1,n+1):
        ng*=i
    kg=1
    for i in range(1,k+1):
        kg*=i
    n_k_g=1
    for i in range(1,n-k+1):
        n_k_g*=i
    return (ng)/(kg*n_k_g)
n,k=4,2
to_hop_chap(k,n)
```

```python
#cah2
def to_hop(n,k):
    if k==n or k==0:
        return 1
    return to_hop(n-1,k)+to_hop(n-1,k-1)
to_hop(4,2)
```
