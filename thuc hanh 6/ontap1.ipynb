```python
#1
def smart_taxi_fare(car_type,star_hour,km,wait_minutes):
    s=0
    if wait_minutes<5:
        w=0
    else:
        if wait_minutes//5==wait_minutes/5:
            w=0
        else:
            w=((wait_minutes//5)+1)*2000
     
    if car_type==4:
        if 0<km<0.8:
            c=11000
        elif 0.8<=km<20:
            c=11000+(km-0.8)*13000
        else:
            c=11000+(20-0.8)*13000+(km-20)*10500
    else:
        if 0<km<0.8:
            c=12000
        elif 0.8<=km<20:
            c=11000+(km-0.8)*14000
        else:
            c=11000+(20-0.8)*13000+(km-20)*12000
    s=c+w

    if int(star_hour) in [22,23,0,1,2,3,4,5]:
        s=s*1.2
    else:
        if km>30:
            s=s*0.9
        if s<50 and km>5:
            s=50000
    return s
smart_taxi_fare(4,7,1,3)

    
```

```python
#6
def five(a,b,c,d,e):
    lis=[a,b,c,d,e]
    for k in lis:
        cou=0
        for i in range(len(lis)):
            if lis[i]==k:
                cou+=1
        if cou>2:
            return True
    return False
five(1,2,2,1,2)
```

```python
#26
ts=lambda a,b,c:(c-a)/(a-b)
def are_collinear(x1,y1,x2,y2,x3,y3):
    if x1==0 and x2==0 and x3==0:
        return True
    elif y1==0 and y2==0 and y3==0:
        return True
    else:
        if ts(x1,x2,x3)==ts(y1,y2,y3):
            return True
        return False
x1,y1=map(int,input().split())
x2,y2=map(int,input().split())
x3,y3=map(int,input().split())
are_collinear(x1,y1,x2,y2,x3,y3)
```
