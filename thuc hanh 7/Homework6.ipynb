```python
# W6A17
n= int(input())
lis=[]
for i in range(n):
    a=tuple(list(map(int,input().split())))
    lis.append(a)
#ma tran tra ra cac tupple moi tup la 1 hang
chinh=[]
phu=[]
for i in range(1,len(lis)+1):
    for j in range(1,len(lis[i-1])+1):
        if i==j:
            chinh.append(lis[i-1][j-1])
        if i+j==n+1:
            phu.append(lis[i-1][j-1])
print(chinh,phu)



```

```python
# W6A21
def dic(x:list)->dict:
    dic={}
    for i in range(len(x)):
        if x[i] in dic:
            dic[x[i]]+=[i]
        else:
            dic[x[i]]=[i]
    return dic

lis=list(map(int,input().split()))
dic(lis)
```

```python
# W6A23
def xoamotgiatri(x:list,y:int)->list:
    lis=[]
    for i in range(len(x)):
        if x[i]>=y:
            lis.append(x[i])
    return lis
xoamotgiatri([1,2,3,5,6,4,1,23,5,3,6,1,2,3,4],3)
```

```python
# W6A24
def dayvesau(x:list)->list:
    a=-len(x)
    dem=-1
    for i in range(-1,a-1,-1):
        if x[i]!=0:
            x[i],x[dem]=x[dem],x[i]
            dem-=1

   
    return x
dayvesau([0, 2, 3, 2, 4 ,0, 0, 0, 9])


```

```python
# W6A2
def ex(x:list)->list:
    a=set(x)
    b=[]
    for i in range(10):
        if i not in a:
            b.append(i)
    return b
ex([1,2,3,4,5,5,1,1,3,2])
```

```python
# W6A26
def s(x:list)->int:
    s=0
    for i in range(1,len(x)-1):
        if x[i]>x[i-1] and x[i]>x[i+1]:
            s+=x[i]
    return s
s([1, 3, 2, 5, 4, 6])

```

```python
# W6A27
def vitri(x:list,y:int)->list:
    lis=[]
    for i in range(len(x)):
        if x[i]<y:
            lis.append(i)
            continue
    if len(lis)==0:
        return -1
    return lis
vitri([1,2,3,4,5,6,7,8,9],-1)
```

```python
# W6A28
def day(x:list)->str:
    if x[0]>x[1]:
        a='giam'
    elif x[0]<x[1]:
        a='tang'
    else:
        return 'vo danh'
    
    for i in range(1,len(x)-1):
        if x[i]<x[i+1]:
            t='tang'
            if a!=t:
                return 'vo danh'
        elif x[i]>x[i+1]:
            t='giam'
            if t!=a:
                return 'vo danh'
        else:
            return 'vo danh'
    return a
day([4,3,2,2,1])



```
