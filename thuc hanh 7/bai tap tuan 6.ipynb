```python
# W6A1
def xoa(a):
    lis=[]
    for i in a:
        if i not in lis:
            lis.append(i)
    return lis
n=list(map(int,input().split()))
xoa(n)
```

```python
# W6A2
def tongtichluy(x):
    lis=[]
    for i in range(len(x)):
        s=0
        for k in range(i+1):
            s+=x[k]
        lis.append(s)
    return lis
x=[1,2,3,4]
tongtichluy(x)
        
```

```python
# W6A4
dic={}
nhap=input().split()
for key in nhap:
    a,b=key.split(':')
    if a in dic:
        dic[a]+=[b]
    else:
        dic[a]=[b]

print(dic)
```

```python
#value là số key là chữ# 
# W6A9
def nhap(a:list):
    dic={}
    for i in a:
        x,y=i.split(":")
        dic[x]=y
    return dic
a=nhap(list(input().split()))
b=nhap(list(input().split()))
for i in a:
    if i in b:
        b[i]=b[i]+int(a[i])
    else:
        b[i]=int(a[i])
b=dict(sorted(b.items()))

print(b)


```

```python
# W6A10
lis=list(map(int,input().split()))
k=int(input())
b=[]
lis.sort()
for x in lis:
    for y in lis:
        if x+y==k:
            if (y,x) in b:
                continue
            else:
                b.append((x,y))
print(b)
```
