```python
def ngto(a):
    if a==1:
        return False
    for i in range(2,int(a**0.5)+1):
        if a%i==0:
            return False
        return True
a=int(input())
ngto(a)
```

```python
def perfec(a):
    s=0
    for i in range(1,a):
        if a%i==0:
            s+=i
    if s==a:
        return True
    return False
a=int(input())
perfec(a)
```

```python
def vitri(lis,k):
    for i in range(len(lis)):
        if lis[i]==k:
            return i
        return -1
lis=list(map(int,input().split()))
k=int(input())
vitri(lis,k)
```

```python
#8
def haming(a,b):
    c=0
    x=a^b
    for k in f"{x:b}":
        if k=='1':
            c+=1
    return c

a,b=101001,11111001
haming(a,b)
```

```python
#10
def dangcau(a,b):
    dic={}
    for i in range(len(a)):
        if a[i] in dic and dic[a[i]]!=b[i]:
            return False
        else:
            dic[a[i]]=b[i]
    return True
a=list(input())
b=list(input())
dangcau(a,b)


    
        
        
```

```python
#12
def divide_with_remainder(dividend, divisor):
    q=dividend//divisor
    r=dividend%divisor
    return q,r
a,b=map(int,input().split())
divide_with_remainder(a,b)
```
